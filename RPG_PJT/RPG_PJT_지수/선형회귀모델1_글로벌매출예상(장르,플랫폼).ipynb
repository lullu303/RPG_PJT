```python
import pandas as pd
game = pd.read_csv('Video_Games(정제2).csv')
game.head()
```

```python
Game = game[['Platform','Genre','Global_']]
Game
```

```python
# 원핫인코딩으로 범주형데이터들 수치화
target = 'Global_'
from category_encoders import OneHotEncoder
encoder = OneHotEncoder(use_cat_names = True)
Game_encoded = encoder.fit_transform(Game)
Game_encoded.head()
```

```python
# 데이터셋 분리
from sklearn.model_selection import train_test_split
# 테스트 분리
trainval, test = train_test_split(Game_encoded)
# 훈련/검증 분리
train, val = train_test_split(trainval)
```

```python
feature = list(train)
print(feature)
```

```python
# X값과 Y값 분리하기
features=['Platform_SNES', 'Platform_N64', 'Platform_X360', 'Platform_PS3', 'Platform_PS', 'Platform_DC', 'Platform_PS4', 'Platform_Wii', 'Platform_XOne', 'Platform_PS2', 'Platform_XB', 'Platform_NES', 'Platform_GC', 'Platform_PC', 'Platform_GB', 'Platform_GBA', 'Platform_3DS', 'Platform_WiiU', 'Platform_PSV', 'Platform_DS', 'Platform_PSP', 'Platform_SAT', 'Platform_GEN', 'Platform_SCD', 'Platform_GG', 'Platform_WS', 'Platform_2600', 'Platform_NG', 'Platform_TG16', 'Platform_3DO', 'Platform_PCFX', 'Genre_Platform', 'Genre_Action', 'Genre_Sports', 'Genre_Fighting', 'Genre_Shooter', 'Genre_Role-Playing', 'Genre_Racing', 'Genre_Strategy', 'Genre_Misc', 'Genre_Adventure', 'Genre_Simulation', 'Genre_Puzzle']
X_train=train[features]
X_val=val[features]
X_test=test[features]
y_train=train[target]
y_val=val[target]
y_test=test[target]
```

```python
# 모델 학습시키기(fit함수)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# 검증 데이터를 통해 모델 정확도 확인해보기
y_val_pred = model.predict(X_val)
```

해당 예측값의 MSE, RMSE, MAE, R2값을 구해보도록하겠다.

MSE: 표준 오차이며 표본 평균 분포의 표준편차를 의미한다. 값은 낮을 수 록 좋다.

RMSE: 평균 제곱근 편차이며 MSE를 루트 씌운 값이다. 값은 낮을 수 록 좋다

MAE: 평균 절대 오차. 값은 낮을 수 록 좋다.


R-squared: 0과 1 사이의 값이 나오며, 1에 가까울 수록 좋은 모델이다.



```python
# 함수이용해서 예측값 구하기
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mse_val=mean_squared_error(y_val, y_val_pred)
rmse_val=mse_val**0.5
mae_val=mean_absolute_error(y_val, y_val_pred)
r2_val=r2_score(y_val, y_val_pred)
print('검증데이터 평가')
print('검증데이터 MSE:', mse_val)
print('검증데이터 RMSE:', rmse_val)
print('검증데이터 MAE:', mae_val)
print('검증데이터 r2:', r2_val)
```

```python
# 검증데이터 학습 시킨 후에 테스트 데이터로 정확도 확인하기
model.fit(X_val, y_val)
y_test_pred=model.predict(X_test)
mse_test=mean_squared_error(y_test, y_test_pred)
rmse_test=mse_test**0.5
mae_test=mean_absolute_error(y_test, y_test_pred)
r2_test=r2_score(y_test, y_test_pred)
print('테스트데이터 평가')
print('테스트데이터 MSE:', mse_test)
print('테스트데이터 RMSE:', rmse_test)
print('테스트데이터 MAE:', mae_test)
print('테스트데이터 r2:', r2_test)
```

r2값이 더 떨어졌다.. 좋은 모델이 아님
