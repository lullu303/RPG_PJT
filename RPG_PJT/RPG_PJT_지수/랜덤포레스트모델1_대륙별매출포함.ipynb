```python
import pandas as pd
df = pd.read_csv('Video_Games(연평균).csv')
df.info()
```

```python
# 이름, 출시년도 변수 제거
df=df.drop(['Name','Year'],axis=1)
# 유저평점 변수 형 변환
df['U_Score'] = pd.to_numeric(df['U_Score'], errors='coerce')
df.head()
```

```python
df.info()
```

```python
# "U_Score" 열의 누락된 값을 최저값으로 채우기
min_score = df["U_Score"].min()
df["U_Score"].fillna(min_score, inplace=True)
```

```python
df.head()
```

```python
# 범주형 변수를 수치형 변수로 변환하기 (원핫인코딩)
df_encoded = pd.get_dummies(df, columns=["Platform", "Genre", "Publisher"])
```

```python
df_encoded.info()
```

```python
# 예측 모델을 위한 독립 변수(X)와 종속 변수(Y) 분리
X = df_encoded.drop("Global_", axis=1)
Y = df_encoded["Global_"]
```

```python
X.head()
```

```python
X.info()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
# 훈련 세트와 테스트 세트로 데이터 분할
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
linear_predictions = linear_model.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
# 랜덤 포레스트 모델 학습 및 예측
rf_model = RandomForestRegressor()
rf_model.fit(X_train, Y_train)
rf_predictions = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# 모델평가
linear_mse = mean_squared_error(Y_test, linear_predictions)
linear_mae = mean_absolute_error(Y_test, linear_predictions)
linear_r2 = r2_score(Y_test, linear_predictions)

rf_mse = mean_squared_error(Y_test, rf_predictions)
rf_mae = mean_absolute_error(Y_test, rf_predictions)
rf_r2 = r2_score(Y_test, rf_predictions)
```

```python
# 평가결과 출력
print("선형 회귀 모델 평가:")
print("MSE:", linear_mse)
print("MAE:", linear_mae)
print("R^2:", linear_r2)
print()

print("랜덤 포레스트 모델 평가:")
print("MSE:", rf_mse)
print("MAE:", rf_mae)
print("R^2:", rf_r2)
```

- 선형회귀모델 : 
    MSE(평균제곱오차) 예측 결과와 실제 결과 간 매우 큰 오차가 있다
    MAE(평균절대오차) 역시 상당히 큰 값
    R2 음수, 모델이 예측하는 것보다 단순히 평균으로 예측하는 것이 더 나은 수준
    : 적합하지 않다.
- 랜덤포레스트 : 
    MSE값 매우 작고 0에 가깝다.
    MAE값 역시 작다
    R2값 0.99로 매우 매우 높다. 
    : 예측 결과와 실제 결과 간에 작은 오차를 가지고 데이터 잘 설명하는 모델

```python
from sklearn.model_selection import cross_val_score
# 랜덤 포레스트 모델에 대한 k-fold 교차검증 수행
cv_scores = cross_val_score(rf_model, X, Y, cv=5, scoring='neg_mean_squared_error')
```

```python
import numpy as np
# 교차 검증 결과 출력
cv_mse = -cv_scores
cv_rmse = np.sqrt(cv_mse)
cv_mean_rmse = np.mean(cv_rmse)

print("교차 검증 결과 RMSE:", cv_rmse)
print("평균 RMSE:", cv_mean_rmse)
```

5 폴드 진행, 예측값과 실제값 사이 오차의 평균은 0.02076 매우(?)작은 편에 속함상당히 정확한 예측을 수행할 가능성이 높다.

```python
import pandas as pd
# 입력해서 사용해보기
# 입력 데이터 생성
info = {'Platform': ['PS4'],
        'Genre': ['Action'],
        'Publisher': ['Electronic Arts'],
        'C_Score': [85]}
input_data = pd.DataFrame(info)
# 모델에 입력 데이터 적용
prediction = rf_model.predict(input_data)
# 예측 결과 출력
print("예측된 판매량:", prediction)
```
