## 주제 : 비디오게임 데이터를 통한 미래시장 예측과 판매 전략수립
 ### => 어떤 장르의 게임을 언제(시대) 어디서(지역, 플랫폼) 출시해야 안정적인 수익을 보장할 수 있을까?

### 1차 목표 : 장르별, 시대별, 대륙별 트렌드와 매출액을 파악하기

- 시대별 장르 선호도 변화는 어떠한가? (시대별 장르 트렌드 변화)
- 어떤 플랫폼에서 게임을 출시해야 하는가? (시대별 플랫폼 판매량)
- 어느 지역에 출시할 것인가? (대륙별 선호도)

### 시대별 게임장르 트렌드

90년대 초반까지는 Sports 장르가 가장 많았지만, 그 이후로부터 2000년대까지 Action 장르의 인기가 급등하면서 Sports 장르를 앞지르게 되었다.

그러나 Sports 역시 2000년까지 꾸준히 증가하는 추세를 보인다.

모든 장르가 1999년 ~ 2009년까지 상승하는 그래프를 보여주지만, 그 이후로부터 전체적으로 감소하는 경향이 나타난다.

Action 장르는 감소하는 기울기가 완만하여 추후 Action 장르의 게임을 출시하더라도 매출에 큰 타격이 없을 것으로 보인다.

두번째 인기 장르인 Sports는 기울기가 가파른 모양을 보이는데, 2010년도에 들어서는 큰 인기가 없어졌음을 알 수 있다.

###  출시된 연도에 출시된 게임 중 그중에 가장 많이 팔린 장르

- 2015년까지는 Action 장르가 가장 많이 팔렸다.
- 2016년에는 Shooter 장르가 가장 많이 팔렸다.

### 연도별 판매량(북미)
 - 출시연도에 따른 판매량
 - 출시연도에 따라서 어떤 장르가 많이 팔렸는가?

- sports,shooter는 초대박 흥행작이 있는 반면 action은 초대박 흥행작이 없다. 
- 초대박 흥행작 없이 action이 가장 많은 수익을 낸 것을 보면 action은 꾸준히 사랑받고 있음을 알 수 있다.

### 연도별 판매량(유럽)

- sports만 초대박 흥행작이 있는걸 봐서, shooter(Duck Hunt)는 북미에서만 인기있었다. 
- 북미와 동일하게 action은 꾸준히 사랑받고 있음을 알 수 있다.

### 연도별 판매량(일본)

- 일본에서 포켓몬 시리즈는 초대박 흥행작으로 상당히 인기있는 게임 시리즈임을 알 수 있다. 
- 포켓몬 외에는 슈퍼마리오가 인기가 있다.

### 연도별 판매량(그외)

- 그 외 대륙에서도 Will는 흥행하였다. 전세계적인 게임임이 확실하다. 
- 그 외 대륙에서는 Action(Grand Theft Auto: San Andreas), Racing(Gran Turismo 4)이 인기가 있다.

### 어느 년도에 어떤 장르가 가장 많이 팔렸는지 보기 쉽게 년도 별로 나눈 그래프

- 2000년대 초 이후로 액션 게임의 판매량이 늘어나는 추세를 볼 수 있다.

- 2000년대 중반 비디오 게임의 판매량이 전체적으로 크게 증가하였다.

- 2010년대 초 액션 게임을 제외한 다른 장르의 게임들의 판매량이 크게 감소하였다.
